#**a) Lakukan Integrasi data pada kedua dataset yaitu dataset diabetes_RS1 dan diabetes_RS2 dengan memperhatikan kolom Age ! (Nb. Penulisan nama kolom harus sama persis, maka ubah salah satu nama kolom terlebih dahulu menjadi Age atau AGE)**

Integrasi data yaitu penggabungan data dari
beberapa data stores (multiple data stores). Integrasi yang cermat dapat membantu mengurangi dan menghindari redundansi dan inkonsistensi dalam kumpulan data yang dihasilkan. \\

Pada contoh kali ini yaitu melakukan penggabungan data dari kedua dataset diabetes seperti berikut

```python
import pandas as pd                         #Kita lakukan import library untuk perhitungan numerik
dataset1 = "/content/diabetes_RS1.csv"      #Read dataset csv yang digunakan
dataset2 = "/content/diabetes_RS2.csv"
df1 = pd.read_csv(dataset1,header = 0)      #Proses pembacaan csv
df2 = pd.read_csv(dataset2,header = 0)
```

untuk menyamankan dari kedua dataset, pertama tama kita cek dahulu atribut kolom yang ada pada dataset diabetes 1


```python
df1.head()
```

cek atribut di dataset 2

```python
df2.head()
```

naah dapat kita lihat dari kedua dataset tersebut, ada satu kolom yang memiliki keterangan dan data yang sama yaitu pada atribut Age(umur). sebelum kita lakukan integrasi kita samakan dulu nama dari atribut tersebut \\

Saya ubah di dataset 1 saja

```python
df1.rename(columns={"Age":"AGE"}, inplace=True)
df1.head()
```

baru kita lakukan integrasi! \\
Sehingga tercipta dataset baru yang saya masukkan dalam variabel df

```python
df = pd.merge(df1, df2, on = 'AGE')
df.head(10)
```

kita save CSV terbarunya!

```python
df.to_csv("data_integration_tugasindv_damin2.csv")
```

#**d) Dari semua dataset yang ada di praktikum kali ini pilihlah minimal satu dataset kemudian lakukan Data Discretization!**
